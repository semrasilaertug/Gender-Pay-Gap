**ABD'de 2015 yılı Ocak ayında çalışan tüm ABD vatandaşlarına dair bir veri çerçevesidir. Bu çalışmanın amacı kadın ve erkekler arasındaki maaş farklarını incelemektir.**

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("inc_occ_gender.csv")
```

```python
data.head()
```

```python
data.columns
```

```python
data.info()
```

```python
butun_sektorler = [sektor for sektor in data.Occupation if sektor.isupper()]
```

```python
butun_sektorler
```

```python
df = data.loc[data['Occupation'].isin(butun_sektorler)]
```

```python
df
```

```python
df.info()
```

```python
df_index = df.set_index(df['Occupation'])
```

```python
df_index
```

```python
df_index.info()
```

```python
del df_index['Occupation']
```

```python
df = df_index
```

```python
df
```

```python
df = df.drop(index='ALL OCCUPATIONS')
df
```

```python
df.dtypes
```

```python
df['All_workers'] = df['All_workers'].astype(int)
df['All_weekly'] = df['All_weekly'].astype(float)
df['M_workers'] = df['M_workers'].astype(int)
df['M_weekly'] = df['M_weekly'].astype(float)
df['F_workers'] = df['F_workers'].astype(int)
df['F_weekly'] = df['F_weekly'].astype(float)
```

```python
df.info()
```

**Kadın ve erkeklerin haftalık maaşlarını görselleştirmekle başlayalım.**

```python
df[['M_weekly','F_weekly']].plot(kind='barh')
```

**Haftalık kazancın kadınlarda daha fazla olduğu bir sektör göremiyoruz.**

**Kadın ve erkeklerin haftalık maaş verilerini kullanarak median_gap sütununu oluşturdum. Bu sütunun işevi kadınların haftalık kazancının erkeklerin haftalık kazancından ne kadar farklı olduğunu yüzde olarak göstermektir.**

```python
df['median_gap'] = (df['F_weekly'] - df['M_weekly']) / df['M_weekly']
```

```python
df['median_gap'].plot(kind='bar')
plt.ylim(None,0.05)
```

**median_gap değerlerinin hepsi negatiftir. Bu sonuç bize hiçbir sektörde kadınların erkeklerden daha fazla maaş almadığını gösterir. Hesaba katmadığımız birçok farklı faktör olabileceğine karşın sonuç hiç de iç açıcı değil.**

**Şimdi de haftalık olarak kadınların maaşlarının erkeklerinkine oranına bakalım. Bu oran 1'e yaklaştıkça cinsiyetler arasındaki 
maaş eşitliği artar.**

```python
df['oran'] = (df['F_weekly'] / df['M_weekly'])
sorted_df_oran = df.sort_values('oran', ascending=True)

plt.figure(figsize=(8, 8))

plt.subplot(2, 1, 1)
plt.barh(range(11), sorted_df_oran.tail(11).oran)
plt.yticks(range(11), sorted_df_oran.tail(11).index, fontsize=10)
plt.plot([1, 1], [0, 11], '--', color='r')
plt.title('Maaş eşitliğinin en yüksek olduğu sektörler')
plt.xlim([0, 1.2])

plt.subplot(2, 1, 2)
plt.barh(range(11), sorted_df_oran.head(11).oran)
plt.yticks(range(11), sorted_df_oran.head(11).index, fontsize=10)
plt.plot([1, 1], [0, 11], '--', color='r')
plt.title('Maaş eşitliğinin en düşük olduğu sektörler')
plt.xlim([0, 1.2])
plt.xlabel('Kadın/Erkek Maaş Oranı')

plt.show()
```

**Görüldüğü üzere hiçbir sektörde cinsiyet bazında maaş eşitliği yoktur.**

**Sektörlere göre kadın çalışanların oranını inceleyelim.**

```python
df['K_calisan_orani'] = (df['F_workers'] / df['All_workers'])
sorted_df_Kcalisan = df.sort_values(['K_calisan_orani'], ascending = [True])

plt.figure(figsize = (10,10))

plt.subplot(2,1,1)
plt.barh(range(11),sorted_df_Kcalisan.tail(11).K_calisan_orani)
plt.yticks(range(11),sorted_df_Kcalisan.tail(11).index, fontsize = 10)
plt.plot([0.5,0.5],[0,11], '--',color = 'r')
#Gösterge çizgisi kadınların ve erkeklerin sektörde %50 oranında yani eşit sayıda varolması düşünülerek çizilir.  
plt.xlim([0,1])
plt.title('Kadın çalışan oranının en yüksek olduğu sektörler')

plt.subplot(2,1,2)
plt.barh(range(11),sorted_df_Kcalisan.head(11).K_calisan_orani)
plt.yticks(range(11),sorted_df_Kcalisan.head(11).index, fontsize = 10)
plt.plot([0.5,0.5],[0,11], '--',color = 'r')
plt.title('Kadın çalışan oranının en düşük olduğu sektörler')
plt.xlim([0,1])
```

**Kadın ve erkek çalışan sayısının eşit olduğu oran 0.5'tir.**

**Kadın ve erkeklerin haftalık maaşlarının toplamının ortalamasını alarak sektörlerin maaş farklarını inceleyelim.**

```python
df['ortak_maas'] = ((df['M_weekly'] + df['F_weekly'])/2)
sorted_df_ortakMaas = df.sort_values(['ortak_maas'], ascending = [True])
'''Kadın ve erkeklerin haftalık aldıkları maaşın ortalamasını alarak cinsiyet faktöründen bağımsız olarak 
hangi sektörlerin yüksek maaşlı hangilerinin düşük maaşlı olduğunu görmeyi amaçladım.'''

plt.figure(figsize = (7,7))

plt.subplot(2,1,1)
plt.barh(range(11),sorted_df_ortakMaas.tail(11).ortak_maas)
plt.yticks(range(11),sorted_df_ortakMaas.tail(11).index, fontsize = 10)
plt.xlim([0,2000])
plt.title('En Yüksek Maaşlı Sektörler')

plt.subplot(2,1,2)
plt.barh(range(11),sorted_df_ortakMaas.head(11).ortak_maas)
plt.yticks(range(11),sorted_df_ortakMaas.head(11).index, fontsize = 10)
plt.xlim([0,2000])
plt.title('En Düşük Maaşlı Sektörler')
```

**Şimdi de maaş farklarını incelediğimiz sektörlerdeki kadın çalışan oranını inceleyelim.**

```python
sorted_df = df.sort_values('ortak_maas', ascending=True)
plt.figure(figsize = (9,9))
plt.subplot(2,1,1)
plt.barh(range(11),sorted_df.tail(11).K_calisan_orani)
plt.yticks(range(11),sorted_df.tail(11).index, fontsize = 10)
plt.plot([0.5,0.5],[0,11], '--',color = 'r')
plt.xlim([0,1])
plt.title('En yüksek maaşlı sektörlerdeki kadın çalışan oranı')

plt.subplot(2,1,2)
plt.barh(range(11),sorted_df.head(11).K_calisan_orani)
plt.yticks(range(11),sorted_df.head(11).index, fontsize = 10)
plt.plot([0.5,0.5],[0,11], '--',color = 'r')
plt.xlim([0,1])
plt.title('En düşük maaşlı sektörlerdeki kadın çalışan oranı')
```

**Bu grafiklerdeki kırmızı çizgiyle belirtilen 0.5 noktası sektörde çalışan kadın ve erkek sayısının eşit olduğunu noktadır.**

**Son olarak sektörlerdeki kadın ve erkeklerin maaşlarını sektörün ortalama maaşı açısından değerlendirelim.**

```python
plt.plot(df.ortak_maas, df.M_weekly,'o')
plt.plot(df.ortak_maas, df.F_weekly,'o')
plt.legend(['Erkek','Kadın'])
plt.xlabel('Sektörlerin ortalama maaşı')
plt.ylabel('Maaş')
plt.show()
```
